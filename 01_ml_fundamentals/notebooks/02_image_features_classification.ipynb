```python
import numpy as np
import matplotlib.pyplot as plt
import random

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

from pathlib import Path
import joblib
```

```python
digits = load_digits()

print(digits.images.shape)

plt.imshow(digits.images[5], cmap="gray")
plt.title(f"Label: {digits.target[5]}")
plt.show()
```

```python
X = digits.images
y = digits.target

# flatten images
X_features = X.reshape(len(X), -1)

print(X_features.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_features, y, test_size=0.35, random_state=42
)

model = RandomForestClassifier()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
idx = random.randint(0, len(X_test))

plt.imshow(X_test[idx].reshape(8,8), cmap="gray")
plt.title(f"Prediction: {y_pred[idx]}")
plt.show()
```

```python
joblib.dump(model, "../results/digit_classifier.pkl")
```
